```python
from pathlib import Path
import scipy.io.wavfile as wavfile
import IPython
import matplotlib.pyplot as plt
```

### Paths

```python
pwd = Path('.')
str(pwd.resolve())
```

```python
sounds = pwd / 'sounds'
```

```python
combined = sounds / 'combined.wav'
str(combined.resolve())
```

### Listen to audio

```python
IPython.display.Audio(str(combined.resolve()))
```

### Import  wav files

```python
for p in sounds.iterdir():
    print(p)
```

```python
def naming(i):
    return (f"samplerate_{i}", f"signal_{i}")
```

```python
signals = [naming(i) for i in range(9)]
```

```python
signals = iter(signals)
```

```python
empty = list()
for wav in sounds.iterdir():
    if wav.match('*.wav'):
        ra, si = next(signals)
        ra, si = wavfile.read(str(wav.resolve()))
        empty.append((ra, si))
```

### Create spectograms

```python
signal_fa = empty[0][1]
plt.specgram(signal_fa);
```

```python
signal_combined = empty[2][1]
plt.specgram(signal_combined)
plt.savefig('plots/combined_spectogram.png');
```
